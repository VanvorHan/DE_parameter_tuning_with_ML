```python
import os
import sys
import pickle
import numpy as np
import matplotlib.pyplot as plt


test_functions = [2, 7, 12, 19, 24]
file_names = ["mut_results.pickle", "pop_results.pickle", "rec_results.pickle"]

for index, fun in enumerate(test_functions):
    print(f"\t********* Function {fun} *********")
    for file_name in file_names:
        if not os.path.isfile(file_name):
            continue
        print(f">>>>>>>>>> {file_name} >>>>>>>>>>>")
        root = pickle.load(open(file_name, "rb"))
        print(root.shape)
        plt.plot(root[index], label=file_name[:3])
    plt.xlabel('Run Iteration')
    plt.ylabel('Error')
    #plt.title("Function "+str(fun))
    plt.legend()
    fig_name = "control-group-function" + str(fun) + ".pdf"
    plt.savefig(fig_name, bbox_inches='tight')
    plt.show()
```
