```python
import os
import sys
import pickle
import numpy as np
import matplotlib.pyplot as plt


thresholds = [0.1, 0.15, 0.2]
test_functions = [2, 7, 12, 19, 24]
params = ["mut", "pop", "rec", "together"]

for param in params:
    for index, fun in enumerate(test_functions):
        print(f"\t********* Function {fun} *********")
        for threshold in thresholds:
            file_name = param + "_results_t" + str(threshold) + ".pickle"
            if not os.path.isfile(file_name):
                continue
            print(f">>>>>>>>>> {file_name} >>>>>>>>>>>")
            root = pickle.load(open(file_name, "rb"))
            plt.plot(root[index,:,1], label=param+"_"+str(threshold))
        plt.xlabel('Run Iteration')
        plt.ylabel('Error')
        plt.title("Function "+str(fun))
        plt.legend()
        fig_name = param + "-function" + str(fun) + ".pdf"
        plt.savefig(fig_name, bbox_inches='tight')
        plt.show()
```
